```python
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
import numpy as np
```

```python
def plot_examples(colormaps):
    """
    Helper function to plot data with associated colormap.
    """
    np.random.seed(19680801)
    data = np.random.randn(10, 10)
    n = len(colormaps)
    fig, axs = plt.subplots(1, n, figsize=(n * 2 + 2, 3),
                            constrained_layout=True, squeeze=False)
    for [ax, cmap] in zip(axs.flat, colormaps):
        psm = ax.pcolormesh(data, cmap=cmap, rasterized=False, vmin=-5, vmax=5)
        fig.colorbar(psm, ax=ax)
    plt.title(r'##30DayChartChallenge 1')  
    plt.annotate('part to whole?', xy=(5, 5), xytext=(5.7, 6.2),
             arrowprops=dict(facecolor='green', shrink=0.01),
             )
    plt.axis('off')
    plt.show()
    

# cmap = ListedColormap(["green","grey",  "lightgrey", "grey", "grey"])
cmap =  ListedColormap(["green", "lightgrey","darkgrey","grey","darkgrey"])
plot_examples([cmap])
```

```python
n = 100
X, Y = np.mgrid[0:n, 0:n]
plt.quiver(X, Y, color="green")
plt.title(r'#30DayChartChallenge Day 01', fontsize=12)

plt.annotate('part to whole?', xy=(25, 25), xytext=(38.7, 39.2),
             arrowprops=dict(arrowstyle="-|>", 
                             connectionstyle="arc3,rad=0.2",
                                  relpos=(1, 0.),
                                  fc="w" )
             )
plt.axis('off')
plt.savefig('day01.png') 
```
